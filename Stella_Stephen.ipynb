# Exploratory Data Visualization

# Univariate Visualization

```python
sns.countplot(x='Age', data=Data)
```

From this graph we can observe that, the people above the age of 26 are more engaged in social media.

```python
plt.rcParams['figure.figsize'] =5,5
labels =Data['Age'].value_counts().index.tolist()
sizes =Data['Age'].value_counts().tolist()
explode = (0, 0.1,0)
colors = ['blue','orange','green','red']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90, textprops={'fontsize': 14})
plt.show()
```

```python
sns.countplot(x='Gender', data=Data)
```

```python
plt.rcParams['figure.figsize'] =5,5
labels =Data['Gender'].value_counts().index.tolist()
sizes =Data['Gender'].value_counts().tolist()
explode = (0, 0.1)
colors = ['blue','orange','green','red']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90, textprops={'fontsize': 14})
plt.show()
```

Here females are mostly using social media.

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
sns.countplot(ax=axes[0, 0], data=Data, x='Social media usage per day', color="m")
sns.countplot(ax=axes[0, 1], data=Data, x='use different social media applications for personal and professional use',color="g")
sns.countplot(ax=axes[0,2], data=Data, x='Quality')
sns.countplot(ax=axes[1,0],data=Data, x ='Preffered social media' )
sns.countplot(ax=axes[1,1],data=Data, x ='Security rating')
plt.show()
```

```python
plt.rcParams['figure.figsize'] =5,5
labels =Data[' still using the same communication app that you did two years ago'].value_counts().index.tolist()
sizes =Data[' still using the same communication app that you did two years ago'].value_counts().tolist()
explode = (0, 0.1,0)
colors = ['blue','orange','green']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90, textprops={'fontsize': 14})
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 10))
sns.countplot(ax=axes[0, 0], data=Data, x='Preffered social media for education & proffession', color="m")
sns.countplot(ax=axes[0, 1], data=Data, x='Convenient',color="g")
sns.countplot(ax=axes[0,2], data=Data, x='Cost Effective')
sns.countplot(ax=axes[1,0],data=Data, x ='Advanced Technology' )
sns.countplot(ax=axes[1,1],data=Data, x ='Better way to communicate for introvert people')
sns.countplot(ax=axes[1,2],data=Data, x='Helps to communicate confidently')
sns.countplot(ax=axes[2,0],data=Data, x='Easy to relay information with large group of people')
plt.show()
```

```python
plt.rcParams['figure.figsize'] =5,5
labels =Data['Security rating for proffessional purpose '].value_counts().index.tolist()
sizes =Data['Security rating for proffessional purpose '].value_counts().tolist()
explode = (0, 0.1,0,0,0)
colors = ['blue','orange','green','red','yellow']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90, textprops={'fontsize': 14})
plt.show()
```

```python
hist_plot = ['Age', 'Gender',
       'Social media usage per day',
       ' still using the same communication app that you did two years ago',
       'use different social media applications for personal and professional use',
        'Quality', 'Preffered social media',
       'Security rating', 'Preffered social media for education & proffession',
       'Convenient', 'Cost Effective', 'Advanced Technology',
       'Better way to communicate for introvert people',
       'Helps to communicate confidently',
       'Easy to relay information with large group of people',
       'Security rating for proffessional purpose ']

fig,ax = plt.subplots(5,3, figsize = (25,25))

for i,j in zip(hist_plot,ax.flatten()):
    sns.distplot(Data[i],ax = j, color="m")
plt.show();
```

# Cor-relation heat map

```python
f, ax = plt.subplots(figsize=(20, 20))
corr =Data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, annot = True)
```
